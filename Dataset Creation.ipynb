# Dataset Creation

This project required 2 types of data - air quality data(pollutant levels) and meteorological data (temperature, humidity, wind speed, etc.) for the capital city of Delhi. 

* **Air Quality Data** - obtained from [this](https://www.kaggle.com/rohanrao/air-quality-data-in-india) Kaggle dataset for Indian Air Quality
* **Weather Data** - obtained from the [Reliable Prognosis Weather Archive](https://rp5.ru/Weather_archive_in_New_Delhi,_Safdarjung_(airport)), for the Safdarjung weather station

The data spans from 01-01-2015 to 30-04-2020.

```python
#Import Libraries Needed
import pandas as pd
```

```python
fileA = pd.read_excel('C:/Users/Anantk/Downloads/city_weather.xls')
```

```python
fileA.head()
```

```python
fileA.columns
```

```python
fileA['Date'] =  pd.to_datetime(fileA['Date'], dayfirst = True)
```

```python
fileA.head()
```

```python
fileB = pd.read_csv('C:/Users/Anantk/Downloads/dataset_2.csv')
```

```python
fileB.head()
```

```python
fileB.columns
```

```python
fileB = fileB.drop(['Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21',
       'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'], axis=1)
```

```python
fileB.head()
```

```python
fileB['Date'] =  pd.to_datetime(fileB['Date'])
```

```python
fileB.head()
```

```python
#merge the 2 files along the Date column
df_merged = pd.merge(fileB, fileA, on='Date')
```

```python
df_merged
```

```python
df_merged.columns
```

```python
df_merged=df_merged.rename(columns={"T": "Temp", "U": "Rel_Humidity", "Ff":"Wind_Speed", "DD": "Wind_Direction"})
```

```python
df_merged = df_merged.drop(['Benzene', 'Toluene', 'Xylene', 'Po', 'P', 'Pa', 'WW'], axis=1)
```

```python
#Exporting the final dataset

df_merged.to_csv('final_dataset.csv', index = False)
```
